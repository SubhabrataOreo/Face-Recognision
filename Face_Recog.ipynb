```python
ret,thresh_binary=cv2.threshold(image_grey,127,255,cv2.THRESH_BINARY)
ret,thresh_binary_inv=cv2.threshold(image_grey,127,255,cv2.THRESH_BINARY_INV)
ret,thresh_trunc=cv2.threshold(image_grey,127,255,cv2.THRESH_TRUNC)
ret,thresh_tozero=cv2.threshold(image_grey,127,255,cv2.THRESH_TOZERO)
ret,thresh_tozero_inv=cv2.threshold(image_grey,127,255,cv2.THRESH_TOZERO_INV)
```

```python
names=['ORIGINAL','THRESH_BINARY','THRESH_BINARY_INV','THRESH_TRUNC','THRESH_TOZERO','THRESH_TOZERO_INV']
images=image_grey,thresh_binary,thresh_binary_inv,thresh_trunc,thresh_tozero,thresh_tozero_inv

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],cmap="gray")
    plt.title(names[i])
    plt.xticks([]),plt.yticks([])
```
